```python
import pandas as pd
import numpy as np
from __future__ import division
import json
```

```python
fName_sports = 'sports.txt'
fName_athletes = 'athletes.txt'
fName_countries = 'countries.txt'
```

```python
df_sports = pd.read_json(json.dumps(json.load(open(fName_sports))['data']))
df_athletes = pd.read_json(json.dumps(json.load(open(fName_athletes))['data']))
df_countries = pd.read_json(json.dumps(json.load(open(fName_countries))['data']))
```

```python
df_athletes.birthdate = pd.to_datetime(df_athletes.birthdate)
df_athletes['country_id']= df_athletes.country.apply(lambda x: x['id'])
df_athletes.country = df_athletes.country.apply(lambda x: x['name'])
df_athletes['sport_id']= df_athletes.sport.apply(lambda x: x['id'])
df_athletes.sport = df_athletes.sport.apply(lambda x: x['name'])
df_athletes['gold'] = df_athletes.medals.apply(lambda x: x['gold'])
df_athletes['silver'] = df_athletes.medals.apply(lambda x: x['silver'])
df_athletes['bronze'] = df_athletes.medals.apply(lambda x: x['bronze'])
df_athletes.medals = df_athletes.gold + df_athletes.silver + df_athletes.bronze
df_athletes['medal_winner'] = df_athletes.medals > 0
```

```python
df_athletes.groupby('medal_winner').age.agg([mean,std])
```

```python
df_athletes[df_athletes.medal_winner].age.hist(normed = True, bins = range(10,50), alpha = 0.5)
df_athletes[~df_athletes.medal_winner].age.hist(normed = True, bins = range(10,50), alpha = 0.5)
```

```python
df_countries['gold_2014'] = df_countries.medals.apply(lambda x: x['gold'])
df_countries['silver_2014'] = df_countries.medals.apply(lambda x: x['silver'])
df_countries['bronze_2014'] = df_countries.medals.apply(lambda x: x['bronze'])
df_countries['medals_2014'] = df_countries.gold_2014 + df_countries.silver_2014 + df_countries.bronze_2014
```

```python
def break_out_historical_medals(medal_hist, medal_type):
    if (medal_hist is not None):
        return medal_hist[medal_type]
        #return pd.DataFrame({'gold_total':medal_hist['gold'], 'silver_total':medal_hist['silver'], 'bronze_total':medal_hist['bronze'], 'medal_total':medal_hist['total']}, index = [0])
    else:
        return np.nan
        #return pd.DataFrame({'gold_total':np.nan, 'silver_total':np.nan, 'bronze_total':np.nan, 'medal_total':np.nan}, index = [0])
```

```python
df_countries['gold_total'] = df_countries.medals_historical_total.apply(lambda x: break_out_historical_medals(x, 'gold'))
df_countries['silver_total'] = df_countries.medals_historical_total.apply(lambda x: break_out_historical_medals(x, 'silver'))
df_countries['bronze_total'] = df_countries.medals_historical_total.apply(lambda x: break_out_historical_medals(x, 'bronze'))
df_countries['medal_total'] = df_countries.medals_historical_total.apply(lambda x: break_out_historical_medals(x, 'total'))
```

```python
k = 0
for i in df_countries[~df_countries.medals_historical_by_sport.isnull()].index.values:
    json_str = df_countries.ix[i].medals_historical_by_sport
    country = df_countries.ix[i]['id']
    # Parse the json and melt it into a frame
    df_temp = pd.read_json(json.dumps(json_str)).transpose().reset_index()
    # Add country
    df_temp['country'] = country
    if k == 0:
        df_hist_by_sport = df_temp
    else:
        df_hist_by_sport = pd.concat([df_hist_by_sport,df_temp], axis = 0)
    k += 1
df_hist_by_sport.columns = ['sport', 'bronze', 'gold', 'silver', 'total', 'country']
```

```python
k = 0
for i in df_countries[~df_countries.medals_historical_by_year.isnull()].index.values:
    json_str = df_countries.ix[i].medals_historical_by_year
    country = df_countries.ix[i]['id']
    # Parse the json and melt it into a frame
    df_temp = pd.read_json(json.dumps(json_str)).transpose().reset_index()
    # Add country
    df_temp['country'] = country
    if k == 0:
        df_hist_by_year = df_temp
    else:
        df_hist_by_year = pd.concat([df_hist_by_year,df_temp], axis = 0)
    k += 1
df_hist_by_year.columns = ['year', 'bronze', 'gold', 'silver', 'total', 'country']
```

```python
k = 0
for i in df_countries[~df_countries.medals_by_sport.isnull()].index.values:
    json_str = df_countries.ix[i].medals_by_sport
    country = df_countries.ix[i]['id']
    # Parse the json and melt it into a frame
    df_temp = pd.read_json(json.dumps(json_str)).transpose().reset_index()
    # Add country
    df_temp['country'] = country
    if k == 0:
        df_by_sport = df_temp
    else:
        df_by_sport = pd.concat([df_hist_by_sport,df_temp], axis = 0)
    k += 1
df_by_sport.drop('index', inplace=True, axis = 1)
```

```python
country_cols = np.append(df_countries.columns[~df_countries.columns.isin(df_countries.filter(like = 'medals').columns.values)].values, ['medals_2014'])
```

```python
df_by_sport.set_index('country').to_csv('df_by_sport.csv')
df_hist_by_sport.set_index('country').to_csv('df_hist_by_sport.csv')
df_hist_by_year.set_index('country').to_csv('df_hist_by_year.csv')
df_athletes.set_index('id').to_csv('df_athletes.csv')
df_sports.set_index('id').to_csv('df_sports.csv')
df_countries[country_cols].set_index('id').to_csv('df_countries.csv')
```
